## Donut Charts

```python
classes = ["Class 3", "Class 4", "Class 5"]
sections = ["3.A", "3.B", "4.A", "4.B", "5.A", "5.B"]
```

```python
class_marks = [45, 40, 35]
section_marks = [45, 40, 40, 35, 25, 20]
```

```python
pie1 = plt.pie(class_marks, radius = 1.3, labels=classes)

pie2 = plt.pie(section_marks, radius=1.3-0.3, labels=sections)

plt.setp(pie1[0], edgecolor='white', width=0.3)
plt.setp(pie2[0], edgecolor='white', width=0.4)
plt.margins(0,0)
```

```python
def create_donut_chart(data1, label1, data2, label2):
    pie1 = plt.pie(data1, radius = 1.3, labels=label1)
    pie2 = plt.pie(data2, radius=1.3-0.3, labels=label2)

    plt.setp(pie1[0], edgecolor='white', width=0.3)
    plt.setp(pie2[0], edgecolor='white', width=0.4)
    plt.margins(0,0)
    plt.show()
```

```python
create_donut_chart(class_marks, classes, section_marks, sections)
```
